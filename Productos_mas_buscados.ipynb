```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('events.csv',low_memory=False)
```

## Productos mas buscados

```python
g = df.loc[df['event']=='searched products']
a = g['search_term'].dropna()
a = a.str.lower()
masBusq = a.value_counts().head(7)
```

```python
x=masBusq
plt1 = x.plot(kind = 'bar',figsize = (10,7),grid=True)
plt1.set_title("Top 7 de productos mas buscados en Trocafone\n",size=24)
plt1.set_ylabel("Cantidad de busquedas",fontsize = 21)
plt1.set_xlabel("Producto buscado",fontsize = 18)
```
