```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('events.csv',low_memory=False)
```

### Verifico que los datos no nulos de la columna search_engine solo estan asociados al evento 'search engine hit'

```python
busq = df.loc[(df['search_engine']).isnull()==False]
```

```python
busq['event'].value_counts()
```

# Top motores de busqueda desde los que personas entraron a Trocafone

```python
x=busq['search_engine'].value_counts()
plt1 = x.plot(kind = 'barh', grid=True,figsize = (11,5))
plt1.set_title("Cantidad de visitas según motor de busqueda\n",size=24)
plt1.set_xlabel("Cantidad de personas",fontsize = 15)
plt1.set_ylabel(" ",fontsize = 18)
plt1
```

```python
plt.xscale('log')
plt2 = x.plot(kind = 'barh', grid=True, figsize = (11,5))
plt2.set_title("Cantidad de visitas según motor de búsqueda (log)\n",size=24)
plt2.set_xlabel("Cantidad de personas (log)",fontsize = 15)
plt2.set_ylabel(" ",fontsize = 18)
```
